```python
import pandas
import sklearn.metrics
ranked = pandas.read_csv("/content/gene_ranks.tsv", header=None, delimiter='\t')
p_vals = pandas.read_excel('/content/SV_Gene_Pairs_Corrected.xlsx', index_col=None, header=0) 
anno = pandas.read_excel('/content/gene_annotation.xlsx', index_col=None, header=0) 
```

```python
anno = anno[['Gene ID', 'Annotation']]
```

```python
gene_list = list(p_vals['Gene'])
```

```python
anno['Gene ID'] = [x[:16] for x in anno['Gene ID']]
```

```python
anno
```

```python
annotation_list = []
for gene in gene_list:
  annot = anno.loc[anno['Gene ID'] == gene].drop('Gene ID', axis = 1)
  if annot.empty:
    annotation_list.append('Blank')
    continue
  annotation_list.append(str(annot.values[0][0]))
```

```python
p_vals['Annotation'] = annotation_list
```

```python
gene_list = list(p_vals['Gene'])
```

```python
rank_list = []
for gene in gene_list:
  smol = ranked.loc[ranked[0] == gene].drop([0,2], axis = 1)
  if smol.empty:
    rank_list.append(-1)
    continue
  rank_list.append(int(smol.values[0])+1)
```

```python
p_vals['Rank'] = rank_list
```

```python
url_index = pandas.read_csv("index.html", header = None, skiprows=9, delimiter=' ')
```

```python
raw_list = list(url_index[16])
url_list = []
for url in raw_list:
  if (isinstance(url, str)):
    url_list.append(url.split('"')[1])
#url_list
```

```python
len(url_list)
```

```python
for url in url_list[0:10]:
    if url == '\n':
        continue
    #x = line.split("\"")
    wget_string = "wget " + url #x[1]
    subprocess.call(wget_string, shell=True)
    print(wget_string)
subprocess.call('gunzip *.gz', shell=True)
```

```python
import glob, os
os.chdir("/content")
fastq_files = []
for file in glob.glob("*.fastq"):
    fastq_files.append(file)
fastq_files
```

```python
#!sudo apt-get install -y fastqc
```

```python
for file in fastq_files:
  fqc = "fastqc -o Output -t 20 --noextract " + file
  print(fqc)
  subprocess.call(fqc, shell=True)
```
